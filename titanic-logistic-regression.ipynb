```python
train = pd.read_csv (r'../input/titanic/train.csv')
test = pd.read_csv (r'../input/titanic/test.csv')
```

```python
Age = train['Age']
Age_new = Age.fillna(Age.mean())
train['Age'] = Age_new
train
```

```python
train_new = train
```

```python
Embarked_new = train_new['Embarked']
Embarked_new.fillna('S')
Embarked_new = pd.get_dummies(Embarked_new)
Embarked_new['Q'] = Embarked_new['Q']*2
Embarked_new['S'] = Embarked_new['S']*3
Embarked_new['C'] = Embarked_new['C'] + Embarked_new['Q'] + Embarked_new['S']
Embarked_new.drop(labels = ['Q','S'], axis = 1, inplace = True)
train_new['Embarked'] = Embarked_new
```

```python
Sex_new = train_new['Sex']
Sex_new = pd.get_dummies(Sex_new)
Sex_new.drop(labels = ['male'],axis = 1, inplace = True)
train_new['Sex'] = Sex_new
```

```python
X = np.array(train_new.drop(['Survived','Name','Ticket','Cabin'],1))
Y = np.array(train_new['Survived'])
X.shape
```

```python
logisticmodel = linear_model.LogisticRegression(solver = 'liblinear')
logisticmodel.fit(X,Y)
```

```python
Predictions = logisticmodel.predict(X)
print(Predictions)
```

```python
logisticmodel.score(X,Y)
```
